```python
# convert("18:14:46", "10/04/24" -> "YYYY-MM-DD HH:MM:SS"
# and add to dataframe
from datetime import datetime

def convert_to_dt(time_str, date_str):
  date_obj = datetime.strptime(date_str, "%m/%d/%y")
  combined_datetime = datetime.combine(date_obj, datetime.strptime(time_str, "%H:%M:%S").time())
  formatted_datetime = combined_datetime.strftime("%Y-%m-%d %H:%M:%S")
  return formatted_datetime

# s = convert_to_dt("18:17:46", "10/21/24")
# print(s)

df['DateTimeStamp'] = df['Date'] + ' ' + df['Time']

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

#df.head(100)
df.head(8)
```
